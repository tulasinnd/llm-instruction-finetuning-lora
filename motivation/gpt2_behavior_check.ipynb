### **GPT-2 Behavior Check**

This notebook is designed to observe the behavior of the **base GPT-2 model**, which has **not been fine-tuned**.  

In this notebook, we will:  
1. Load the pretrained GPT-2 model.  
2. Test it with a few sample instructions.  
3. Analyze and record GPT-2’s responses to understand its current behavior and limitations.  

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import warnings
warnings.filterwarnings("ignore")

# Load tokenizer + model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
model.eval()

# List of sample prompts
prompts = [
"what is sun?",
"Explain what is the sun in simple words.",
"Q: What is the sun? A:",
"The sun is a star that",
"What is sun? What is moon?",
"Once upon a time"
]

# Loop over prompts and generate outputs
for prompt in prompts:
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(
        **inputs,
        max_length=50,
        do_sample=True,       # adds randomness
        temperature=0.9,
        top_k=50,
        pad_token_id=tokenizer.eos_token_id  # prevents the "setting pad_token_id.." message
    )
    result = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print(f"Prompt: {prompt}")
    print(f"Response: {result}")
    print("-" * 50)
```

### **Conclusion**

As expected, the base GPT-2 model **does not follow instructions properly**. From the examples above, we notice several issues:

* Struggles to follow clear instructions (like “explain in simple words”).  
* Very sensitive to how the prompt is written — small changes can affect answers a lot.  
* Sometimes repeats phrases or goes in loops.  
* Copies text patterns instead of reasoning.  
* Mixes correct and incorrect information.  
* Works better for continuing text, like stories.

**How to fix this:**  
We can improve GPT-2 using **instruction fine-tuning**. This means training the model on a set of **instruction-response examples** so it learns to follow instructions better. Techniques like **LoRA** let us do this efficiently without retraining the whole model. After this, GPT-2 will give more accurate, relevant, and consistent answers when we ask it tasks.
